 # Citi Bike Data Cleanse


```python
# import pandas library

import pandas as pd
```

```python
# read in files to be cleaned


df3 = pd.read_csv("Resources/202307-citibike-tripdata.csv", low_memory=False)
df2 = pd.read_csv("Resources/202304-citibike-tripdata.csv", low_memory=False)
df1 = pd.read_csv("Resources/202301-citibike-tripdata.csv", low_memory=False)


```

```python
# Create and save new csv file of combined data

frames = [df1, df2, df3]
df = pd.concat(frames).drop_duplicates()
# df.to_csv("Resources/citibike_last_year.csv", index=False)
```

```python
len(df)
```

```python
# Remove any rows with missing data (removes nulls and blank cells)


df[df['ride_id'].astype(bool)]                                                                                                                  
df[df['rideable_type'].astype(bool)]                                                                                                                  
df[df['started_at'].astype(bool)]                                                                                                                  
df[df['ended_at'].astype(bool)]                                                                                                                  
df[df['start_station_id'].astype(bool)]                                                                                                                  
df[df['end_station_id'].astype(bool)]                                                                                                                  
df[df['member_casual'].astype(bool)]                                                                                                                  
df[df['start_station_name'].astype(bool)]                                                                                                                  
df[df['end_station_name'].astype(bool)]                                                                                                                  
df[df['start_lat'].astype(bool)]                                                                                                                  
df[df['start_lng'].astype(bool)]                                                                                                                  
df[df['end_lat'].astype(bool)]                                                                                                                  
df[df['end_lng'].astype(bool)]                                                                                                                  
                                                                                                                
```

```python
# strip out trips that start and end at the same location
df = df.loc[((df["start_station_id"] != df["end_station_id"])),:]
len(df)
```

```python
# Save station data to strip it out of citibike file

# get all unique station data for starting stations
stations_df = df[["start_station_id","start_station_name", "start_lat", "start_lng"]].drop_duplicates("start_station_id")
stations_df.columns = ['station_id', 'station_name', 'lat', 'lng']

# get all unique station data for ending stations
end_df = df[["end_station_id","end_station_name", "end_lat", "end_lng"]].drop_duplicates("end_station_id")
end_df.columns = ['station_id', 'station_name', 'lat', 'lng']

# combine station data and remove duplicates and drop any rows containing nulls
stations_df = pd.concat([stations_df, end_df]).drop_duplicates('station_id')
stations_df = stations_df.dropna()
stations_df.to_csv("Resources/cleaned/citibike_stations.csv", index=False)
```

```python
stations_df
```

```python
# Clear all station data except start and end station id's
df_small = df.drop(columns=['start_station_name', 'end_station_name', 'start_lat', 'start_lng', 'end_lat', 'end_lng'])
len(df_small)
```

```python
df_small.head()
```

```python
#  Convert started_at and ended_at to date time
df_small[["started_at", "ended_at"]] = df_small[["started_at", "ended_at"]].apply(pd.to_datetime)
```

```python
# Calculate trip length in minutes and add to df
df_small['trip_length_min'] = (df_small.ended_at - df_small.started_at) / pd.Timedelta(minutes=1)
df_small.head()
```

```python
# Remove any rides that are less than a minute
df_small = df_small.loc[(df_small["trip_length_min"]>1),:]
len(df_small)
```

```python
df_small.to_csv("Resources/cleaned/citibike_ly_clean.csv", index=False)
```
